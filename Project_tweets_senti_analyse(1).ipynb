```python
import pandas as pd
dt = pd.read_csv('/content/[SubtitleTools.com] 567.csv', sep=';')
dt = dt.dropna()
#Chose first 200 rows
test = dt.iloc[0:200]
test
```

```python
#Make some text preprocessing
import numpy as np
headlines_array = test["Header"]
headlines_list = list(headlines_array)
print(headlines_list)
print(len(headlines_list))
```

```python
# Download nesessary packages
!pip install transformers
from transformers import AutoTokenizer, AutoModelForSequenceClassification

#In this research wi use the BERT model named "ProsusAI/finbert" - download it
tokenizer = AutoTokenizer.from_pretrained("ProsusAI/finbert")
model = AutoModelForSequenceClassification.from_pretrained("ProsusAI/finbert")
```

```python
# Vectorise our data
inputs = tokenizer(headlines_list, padding = True, truncation = True, return_tensors='pt')
print(inputs)
```

```python
#Put our data (Headers) to the model
outputs = model(**inputs)
print(outputs)
```

```python
import torch
#Through the torch functions got the prediction of  probability for each tone 
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

```python
#For easies understanding of probabilities make the labels: Negative, Positive and Neutral
output = open("results_ch", 'w')
import pandas as pd
positive = predictions[:, 0].tolist()
negative = predictions[:, 1].tolist()
neutral = predictions[:, 2].tolist()

table = {'Headline':headlines_list,
         'Date':test["Date"],
         'Company':test["Company"],
         'Change':test["Change"],
         "Positive":positive,
         "Negative":negative, 
         "Neutral":neutral}
      
dk = pd.DataFrame(table, columns = ["Headline", "Date", "Company", "Change", "Positive", "Negative", "Neutral"])
# Сброс ограничений на количество выводимых рядов
pd.set_option('display.max_rows', None)
 
# Сброс ограничений на число столбцов
pd.set_option('display.max_columns', None)
 
# Сброс ограничений на количество символов в записи
pd.set_option('display.max_colwidth', None)
dk
print(dk,file=output)
output.close()
```

```python
#Add the helpfull column "Result" which describes the result of predictions
dk['Result'] = 0
#Make the conditions for labeling News as 1(Positive is the max), 0(Neutral is the max), -1 (Negative) is the max), 
dk['Result1'] = np.where((dk.Negative > dk.Positive)&(dk.Negative > dk.Neutral), '-1', '0')
dk['Result'] = np.where((dk.Positive > dk.Negative)&(dk.Positive > dk.Neutral), '1', dk['Result1'])
#Delete the unnessesary column
dk = dk.drop('Result1', 1)
dk
```

```python
#Goup all headers via company's name and Tweet's date
output = open("/content/Result.csv", 'w')
dk1 = dk.sort_values(
 by=["Company", "Date"]
)
print(dk1, file=output)
output.close()
```
